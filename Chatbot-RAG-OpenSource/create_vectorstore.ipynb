## Fragmentando texto

----------------------------------------

```python
fragments = fh.fragments(obj=pages, chunk_size=384, debug=True)
```

----------------------------------

## Creando vectorstore

```python
from langchain.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings()
print(embedding.client)

path_index = 'G:/Mi unidad/Cursos/Platzi/Chatbot RAG/'
name_index = 'indice-RAG-Hemograma'

vectorstore = Chroma.from_documents(documents=fragments, embedding=embedding, persist_directory = path_index+name_index)
```

-------------------------------------

## Creando retriever

```python
retriever = vectorstore.as_retriever(search_type="similarity",search_kwargs={'k': 10})
```

```python
retriever.invoke('hemoglobina')
```
